```python
import pandas as pd
```

```python
data1 = {
  "name": ["anmol","prakul","vineet"],
  "age": [20,21,19],
  "stream":["data science","data science","data science"]
}

data2 = {
  "name": ["anshul", "aashutosh", "pritam"],
  "age": [24, 21, 22],
  "stream":["aws","aws","aws"]
}

df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
newdf = df1.merge(df2, how='outer') # The merge() method updates the content of two DataFrame by merging them together
```

```python
newdf
```

```python
newdf = df1.merge(df2, how='left')
```

```python
newdf
```

```python
newdf = df1.merge(df2, how='right')
```

```python
newdf
```

```python
newdf = df1.merge(df2, how='cross')
```

```python
newdf
```

```python
# concating two data frames
newdf=df1,df2
```

```python
newdf
```

```python
newdf=pd.concat(newdf)
```

```python
newdf.reset_index()
```

```python
data1 = {
    "id":[101,102,103],
  "name": ["anmol","prakul","vineet"],
  "age": [20,21,19],
  "stream":["data science","aws","developer"]
}

data2 = {
    "id":[101,102,103],
  "project": ["attretion", "cloud computing", "pattern designer"],
  "date":["11/02/2025","21/02/2025","22/01/2025"]
}

df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
newdf = df1.merge(df2, how='right',on="id")
```

```python
newdf
```

```python
b=[df1,df2]
new_df2=pd.concat(b)
new_df2.reset_index()
```

```python
new_df2=pd.concat([df1,df2],axis=1)
new_df2.reset_index()
```

```python
result = pd.concat([df1, df2], axis=1, ignore_index=True)
result
```
