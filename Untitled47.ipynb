```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/covid_toy.csv")
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer()
```

```python
df["fever"] = si.fit_transform(df[["fever"]])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df["gender"] = lb.fit_transform(df["gender"])
df["cough"] = lb.fit_transform(df["cough"])
df["city"] = lb.fit_transform(df["city"])
df["has_covid"] = lb.fit_transform(df["has_covid"])

```

```python
x = df.drop(columns = "has_covid")
y = df["has_covid"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2 , random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(x_train,y_train)
```

```python
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# create the perameter grifd based on the result of random search
params = {
    "criterion" : ["gini","entropy"],
    "max_depth" :[2,3,5,10,20],
    "min_samples_leaf" : [5,10,20,50,100]
}
```

```python
# instance the grid search model
grid_search = GridSearchCV(estimator = dt,param_grid = params,cv = 6
                           , verbose = 1 , n_jobs = -1 , scoring= "accuracy")
```

```python
grid_search = grid_search.fit(x_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
y_pred_2 = grid_search.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred_2)
```
