```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/Attrition.csv")
```

```python
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
df.isnull().sum()
```

```python
x = df.drop(columns = ['Attrition'],axis = 1)
y = df['Attrition']
```

```python
categorical_columns = ['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime','Over18']
```

```python
x_encoded = pd.get_dummies(x,columns = categorical_columns, drop_first = True,dtype=int)
```

```python
# encode the target
lb = LabelEncoder()
y_encoded = lb.fit_transform(y)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x_encoded,y_encoded,test_size = 0.2,random_state = 42)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
df.head()
```

```python
# now we will create architecture of our model
```

```python
import tensorflow as tf

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(units=128, activation='relu', input_shape=(x_train.shape[1],)),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(units=64, activation='relu'),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(units=1, activation='sigmoid')  # For binary classification
])
```

```python
loss_fn = tf.keras.losses.BinaryCrossentropy()
metrices=['accuracy']
```

```python
learning_rate = 0.001
momentum = 0.9
optimizer = tf.keras.optimizers.SGD(learning_rate=learning_rate, momentum=momentum,nesterov=True)
```

```python
model.compile(optimizer=optimizer, loss=loss_fn, metrics=metrices)
```

```python
# training the model
```

```python
model.fit(x_train,y_train,epochs=100,batch_size=8,validation_split=0.1)
```

```python
loss , accuracy =model.evaluate(x_test,y_test)
```

```python
print(f"Loss : {loss}")
print(f"Accuracy : {accuracy}")
```
