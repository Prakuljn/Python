```python
# so , now we will replacce the value of sex and smoker columns with 1 and 0 as both these columns comtain string values
```

```python
df["sex"] = df['sex'].map({'female': 0, 'male': 1})
df["smoker"] = df['smoker'].map({'no': 0, 'yes': 1})
```

```python
# now lets have a at the distribution of the region where people are living according to the dataset
```

```python
import plotly.express as px
```

```python
pie = df ['region'].value_counts()
regions = pie.index
population = pie.values
fig = px.pie(values = population , names = regions , title = "Regions")
fig.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 300],
    'max_depth': [None, 10, 30],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': ['sqrt', 'log2']
}
```

```python
rf = RandomForestRegressor(random_state=42)
```

```python
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid,
                           cv=5, n_jobs=-1, scoring='neg_mean_squared_error')
```

```python
grid_search.fit(x_train, y_train)
```

```python
print(grid_search.best_params_)
```
